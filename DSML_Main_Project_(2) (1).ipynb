# **Feature Engineering**

```python
import pandas as pd
import numpy as np


```

```python
# Convert time column

df['Transaction_Time'] = pd.to_datetime(df['Transaction_Time'], format='%H:%M:%S')
```

**`# Time features`**

```python
df['hour'] = df['Transaction_Time'].dt.hour
df['day_of_week'] = df['Transaction_Time'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].isin([5,6]).astype(int)
```

**`# Card behavior`**

```python
df['avg_amount'] = df.groupby('Customer_ID')['Purchase_Amount'].transform('mean')
df['amount_diff'] = df['Purchase_Amount'] - df['avg_amount']
```

**`# Velocity`**

```python
df = df.sort_values(['Customer_ID','Transaction_Time'])
df['time_diff_min'] = df.groupby('Customer_ID')['Transaction_Time'].diff().dt.total_seconds()/60
df['time_diff_min'] = df['time_diff_min'].fillna(df['time_diff_min'].median())
```

**`# Merchant stats`**

```python
merchant_fraud = df.groupby('Store_ID')['Fraud_Flag'].mean()
df['merchant_risk'] = df['Store_ID'].map(merchant_fraud)
```

**`# Anomaly flags`**

```python
df['is_high_amount'] = (df['Purchase_Amount'] > df['Purchase_Amount'].median()*2).astype(int)
df['is_night'] = ((df['hour'] < 6) | (df['hour'] > 22)).astype(int)
```

# **Feature Selection**



```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.impute import SimpleImputer
```

**# 2. Data Preparation and Cleaning**

```python
# Identify the target variable
TARGET_VARIABLE = 'Fraud_Flag'
y = df[TARGET_VARIABLE]
```

```python
# Identify numerical features for ANOVA
numerical_features = ['Customer_Age', 'Purchase_Amount', 'Footfall_Count']
X_numerical = df[numerical_features].copy()
```

```python
# Handle missing values in 'Customer_Age' by imputing with the mean
imputer = SimpleImputer(strategy='mean')
X_numerical_imputed = pd.DataFrame(
    imputer.fit_transform(X_numerical),
    columns=numerical_features
)
```

**# 3. Feature Selection using ANOVA-based F-test (f_classif)**

```python
# We will select the top 2 best numerical features (adjust k as needed)
k_best = 2
selector = SelectKBest(score_func=f_classif, k=k_best)
X_selected = selector.fit_transform(X_numerical_imputed, y)
```

```python
# Get the names of the selected features
selected_indices = selector.get_support(indices=True)
selected_features = X_numerical_imputed.columns[selected_indices].tolist()
```

```python
print(f"\n--- ANOVA Feature Selection Results ---")
print(f"All numerical features: {numerical_features}")
print(f"Selected {k_best} features using ANOVA F-test: {selected_features}")
```

```python
# Create the final feature DataFrame with only the selected features
X = pd.DataFrame(X_selected, columns=selected_features)
```

**# 4. Train-Test Split**

```python
# Split the data (using a common 80/20 split)
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,  # 20% for testing
    random_state=42, # for reproducibility
    stratify=y       # crucial for imbalanced data like fraud detection
)
```

```python
print(f"\n--- Train/Test Split Details ---")
print(f"Original df size: {len(df)}")
print(f"Training set size: {len(X_train)} ({len(X_train) / len(df) * 100:.2f}%)")
print(f"Testing set size: {len(X_test)} ({len(X_test) / len(df) * 100:.2f}%)")
print(f"Features used in the split: {X_train.columns.tolist()}")
```
